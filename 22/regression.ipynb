```python
x = df['petal_length']
y = df['petal_width']
mean_x = df['petal_length'].mean()
mean_y = df['petal_width'].mean()

diff_y = 0

diff_x = 0
x = x.array
y = y.array
for i,x_i in enumerate(x):
    diff_x += (x_i-mean_x)*(y[i]-mean_y)
    diff_y += (x_i-mean_x)*(x_i-mean_x)
    
A = diff_y/(diff_x)
B = (mean_y) - (A*mean_x)
A,B
```
